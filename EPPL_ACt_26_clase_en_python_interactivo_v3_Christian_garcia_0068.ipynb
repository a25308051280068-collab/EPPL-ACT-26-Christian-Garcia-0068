```python
class Evento:
    def __init__(self):
        # Inicializamos los atributos vacíos
        self.nombre = ""
        self.fecha = ""
        self.lugar = ""

    def leer_datos(self):
        print("\n--- Introduciendo datos del evento ---")
        self.nombre = input("Ingresa el nombre del evento: ")
        self.fecha = input("Ingresa la fecha (ej. 12/12/2024): ")
        self.lugar = input("Ingresa el lugar: ")

    def mostrar_datos(self):
        print("\n--- Información del Evento ---")
        if self.nombre:
            print(f"Evento: {self.nombre}")
            print(f"Fecha:  {self.fecha}")
            print(f"Lugar:  {self.lugar}")
        else:
            print("No hay datos registrados para este evento.")

# --- Creación de Objetos y Lógica Interactiva ---

# Creando el primer objeto (evento1)
evento1 = Evento()
print("Configurando el Evento 1:")
evento1.leer_datos()

# Creando el segundo objeto (evento2)
evento2 = Evento()
print("\nConfigurando el Evento 2:")
evento2.leer_datos()

# --- Resultados ---
print("\n" + "="*30)
print("RESUMEN DE EVENTOS REGISTRADOS")
print("="*30)

evento1.mostrar_datos()
evento2.mostrar_datos()
```
